## 1. 엑셀에 저장한 실제 데이터를 불러온다.
## 2. 단어별로 쪼갠다.
## 3. 단어를 카운팅한다.


```python
# 엑셀을 읽기 위한 준비과정
from openpyxl import load_workbook

read_wb = load_workbook('./bluehouse.xlsx')
read_ws = read_wb.active

list_excel = []

for i in range(1,151):
    print(read_ws.cell(i,1).value)
    list_excel.append(read_ws.cell(i,1).value)
```

```python
list_excel
```

## 가공을 위해 자연어처리 패키지(konlpy.tag), collections 사용

```python
from konlpy.tag import Kkma
import collections
```

```python
kkma = Kkma() # 자연어처리 선언
```

```python
list_temp = []
```

```python
for row in list_excel:
    print(kkma.nouns(row)) #한줄씩 단어별로 쪼갬.
    list_temp = list_temp + kkma.nouns(row)
```

```python
print(list_temp)
```

## 한글자는 단어라 보기 힘들기 때문에 카운팅에 포함시키지 않는다.


```python
result_list = []
```

```python
for check in list_temp:
    print(check)
    if len(check) > 1:
        result_list.append(check)
```

```python
result_list
```

## 청와대 국민청원 게시판 가장많이 검색된 단어

```python
collections.Counter(result_list).most_common(1)
```

## 청와대 국민청원 게시판 Top keyword 5개

```python
collections.Counter(result_list).most_common(5)
```
